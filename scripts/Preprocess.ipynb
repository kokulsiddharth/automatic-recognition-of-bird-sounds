# Preprocess Audio Files


*   This script helps to split the audio file with birds calls into invidual recordings (e.g. if a recording consists of 5 calls by a Koel. This script helps to split the recording into 5 recordings such that each call by the Koel .will be an audio file) 

*   This is done by identifying blocks of energy in the audio clip and saving them as individual files



```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import librosa, librosa.display
from glob import iglob
from scipy.io import wavfile
```

```python
def get_data(file,total):
    """ This function reads names of wav files in input folder and returns a list """
    lst = []
    count = 0 
    for item in iglob(file + '/*.wav'):
        if count >= total:
            break
        else:
            lst.append(item)
            count += 1
        
    lst.sort()
    return lst
```

```python
def get_energy(item):
    """ This function calculates the energy of the audio file and returns the timings of the frames, energy and audio sample values(44.1 Khz)"""
    #Load audio files 
    # sample_rate, samples = wavfile.read(item)
    sampling_rate = 44100
    samples, sample_rate = librosa.load(item, sr = sampling_rate, mono=False)
    # if (len(samples.shape) > 1 and samples.shape[1] == 2):
    #   samples = samples[:,0]
    if (len(samples.shape) > 1 and samples.shape[0] == 2):
        samples = samples[0,:]

    # x = samples[:10*sample_rate]
    x = samples 
    #define frame and hop length 
    frame_length = 512*2
    hop_length = 256*2

    #find energy of audio clip 
    energy = np.array([sum(abs(x[i:i+frame_length]**2)) for i in range(0, len(x), hop_length)])
    frames = range(len(energy))
    t = librosa.frames_to_time(frames, sr=sample_rate, hop_length=hop_length)
#     plt.figure
#     plt.plot(t,energy)
#     plt.title("Energy vs Time")
#     plt.axhline(y=0.7, color='r', linestyle='--')
    return t,energy,x
```

```python
def separate_energy(energy,threshold):
    """ This function separates and returns indexes of blocks of energy"""
    
    #Making energy values below the threshold 0
    adjusted_energy = np.where(energy < threshold,0,energy)
    #Find index of non-zero energy values 
    index = np.nonzero(adjusted_energy)[0]

    energy_blocks = []
    temp = []

    for i in range(1,len(index)):
        if (index[i] - index[i-1] == 1):
            if i == 1:
                temp.append(index[0])  
            #adding index of consecutive energy values in temp list 
            temp.append(index[i]) 
            #adding last block of energy values energy_blocks list  
            if (i == len(index)-1 and len(temp) != 0):
                energy_blocks.append(temp)
                temp = []   
        else:
            #adding energy block to energy_blocks list
            energy_blocks.append(temp)
            temp = []
            temp.append(index[i])

    return energy_blocks 
```

```python
# save(samples,energy_split,energy,time,name)
def save(x,lst,energy,t,name):
    """ This function pads the blocks of energy to ensure they are the same length and saves 20% of the testing, 20% for validation and 60% for training the model"""
    sampling_rate = 44100
   
    #Filter blocks of energy longer than 0.4 seconds  
#     lst = list(filter(lambda x: t[x][-1] - t[x][0] > 0.4, lst))
    lst = [i for i in energy_split if i and time[i][-1] - time[i][0] > 0.4]
    
    # plot different blocks of energy 
#     plt.figure()
#     for i in lst:
#         plt.plot(t[i],energy[i])
#         plt.title("Energy Split vs Time")
    
    #plot original wav file 
    # plt.figure()
    # time = np.linspace(0.0,len(x)/sampling_rate,len(x))
    # plt.title("Original Signal")
    # plt.plot(time,x) 

    #pad and save blocks of energy as wav and npy files 
    pad_length = 122080//2
    pad_type = ["left", "both", "right"]

    for pad in pad_type:
        train_arr = []
        test_arr = []
        val_arr = []
        sample_lst = []
        count = 1
        # plt.figure()
        for m in lst:
            samp = np.arange(int(t[m][0]*sampling_rate),int(t[m][-1]*sampling_rate),1)
            tm = np.linspace(t[m][0],t[m][-1],len(samp))
            # plt.plot(tm,x[samp])
            # plt.title("Split Signal")
            orig_samples = x[samp]
            if len(orig_samples) < pad_length:
                diff = pad_length-len(orig_samples) 
                if pad == "left":
                    orig_samples = np.pad(orig_samples, (diff,0), 'constant', constant_values=(0, 0))
                elif pad == "both":
                    if diff%2 == 0:
                        orig_samples = np.pad(orig_samples, (int(diff/2),int(diff/2)), 'constant', constant_values=(0, 0))
                    else:
                        orig_samples = np.pad(orig_samples, (int(diff/2)+1,int(diff/2)+1), 'constant', constant_values=(0, 0))
                        orig_samples = orig_samples[:pad_length] 
                elif pad == "right":
                    orig_samples = np.pad(orig_samples, (0,diff), 'constant', constant_values=(0, 0))
                sample_lst.append(orig_samples)
            else:
                for i in range(0,len(orig_samples),pad_length):
                    samples_edited = samples[i:i+pad_length]
                    if len(samples_edited) >= pad_length:
                        sample_lst.append(samples_edited[:pad_length])
        
        calls = len(sample_lst)
        print('Calls', calls)
        test_count = round(0.2 * calls)
        validation_count = test_count*2
        
        if pad == "right":
#             for idx,ele in enumerate(sample_lst[:test_count]):
#                 wavfile.write(test_audio + '/' + name + '_' + str(idx) + '_' + pad +'_test.wav', sampling_rate, ele)
                            
            with open(test_npy + '/' + name + "_" + pad + "_" + 'tst.npy', 'wb') as f:
                test_arr = np.array(sample_lst[:test_count])
                np.save(f, test_arr)
                
            with open(validation_npy + '/' + name + '_' + pad + '_' + 'val.npy', 'wb') as f:
                val_arr = np.array(sample_lst[test_count:validation_count])
                np.save(f, val_arr)
            
            with open(train_npy + '/' + name + '_' + pad + '_' + 'trn.npy', 'wb') as f:
                train_arr = np.array(sample_lst[validation_count:]) 
                np.save(f, train_arr)
            
            print("Pad type: {}, test: {}, val: {}, train: {}".format(pad,test_arr.shape,val_arr.shape,train_arr.shape))  
        else:
            with open(train_npy + '/' + name + '_' + pad + '_' + 'trn.npy', 'wb') as f:
                train_arr = np.array(sample_lst[validation_count:]) 
                np.save(f, train_arr)
            print("Pad type: {}, train: {}".format(pad,train_arr.shape))     
        
    return test_count,validation_count-test_count,(calls-validation_count)*3
    
```

```python
audio_data = '../bird-sounds-data/birds_audio' 

validation_audio = '../bird-sounds-data/validation/audio'
train_audio = '../bird-sounds-data/train/audio'
test_audio = '../bird-sounds-data/test/audio'

validation_npy = '../bird-sounds-data/validation/arrays'
train_npy = '../bird-sounds-data/train/arrays'
test_npy = '../bird-sounds-data/test/arrays'
```

```python
test_count_lst = []
train_count_lst = []
validation_count_lst = []

birds_threshold = {'asian_koel': 0.1, 'blue_jay': 0.001, 'cockatoo': 0.05, 'crane': 0.001, 'crow': 0.1, 'cuckoo': 0.01, 'dove': 0.01, 'duck': 0.01, 'eagle': 0.004, 
                   'falcon': 0.0075, 'goose': 0.1, 'grackles': 0.0025, 'hawk': 0.1, 'hen': 0.005, 'humming_bird': 0.001, 'kingfisher': 0.006, 'loon': 0.005, 'ostrich': 0.005, 
                   'parrot': 0.0053, 'penguin': 0.045, 'pigeon': 0.0025, 'quail': 0.0009, 'red_cardinal': 0.012, 'seagull': 0.01, 'sparrow': 1e-05, 'starling': 0.0015, 'swan': 0.01, 
                   'tufted_titmouse': 0.001, 'turkey': 0.01, 'wood_thrush': 0.0005}


data = get_data(audio_data,30)

for ele in data:
    name = ele[len(audio_data)+1:-4]
    time,energy,samples = get_energy(ele)
    # plt.figure
    # plt.plot(time,energy)
    # plt.title("Energy vs Time")
    # plt.axhline(y=0.001, color='r', linestyle='--')
    threshold = birds_threshold[name]
    energy_split = separate_energy(energy,threshold)
    test_count,validation_count,train_count = save(samples,energy_split,energy,time,name)
    test_count_lst.append(test_count)
    train_count_lst.append(train_count)
    validation_count_lst.append(validation_count)
    print(name, 'test_count:', test_count, 'validation_count:', validation_count,'train_count:', train_count) 

print('total_train_count', sum(train_count_lst))
print('total_test_count', sum(test_count_lst))
print('total_validation_count', sum(validation_count_lst))
```
