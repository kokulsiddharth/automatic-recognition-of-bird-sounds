```python
def get_melspectrogram(input_file_dir, output_file_dir):
    """ This function plots the spectrogram of the audio files"""
    spect_data = {}
    count_data = []
    count = 0 
    spect_lst= []
    
  
    for file in iglob(input_file_dir + '/*.npy'):
      bird_spect_lst = []
      name = file[len(input_file_dir)+1:-14]
      with open(file, 'rb') as f:
        audio_data = np.load(f)
      
      for i in range(audio_data.shape[0]):
        samples = audio_data[i]
        mel_spec = librosa.feature.melspectrogram(samples, n_fft=WINDOW_LEN, hop_length=OVERLAP, n_mels=N_MELS,
                                                  sr=SAMPLING_RATE, power=1.0, fmin=FMIN, fmax=FMAX)          
        mel_spec_db = librosa.amplitude_to_db(mel_spec, ref=np.max)        
        mel_spec_db = np.clip((mel_spec_db - np.amin(mel_spec_db)) / -np.amin(mel_spec_db), 0, 1)
        
        # plt.figure()
        # librosa.display.specshow(mel_spec_db, sr=sampling_rate, hop_length=overlap, x_axis='time', y_axis='mel',fmin=fmin, fmax=fmax)
        # plt.colorbar(format='%+2.0f dB')
        # plt.title(item[len(test) + 1 :-4] + '.png')
        # plt.tight_layout()
        # plt.savefig(test_output + '/' + "mel_" + item[26:-4] + '.png')
        # print(item[len(test) + 1 :-4] + '.png')

        mel_spec_db = np.reshape(mel_spec_db, (mel_spec_db.shape[0],mel_spec_db.shape[1],1))
        spect_lst.append(mel_spec_db)
        bird_spect_lst.append(mel_spec_db)

      if ("train" in input_file_dir) and ("train" in output_file_dir):
        model_type = "train"
      elif ("test" in input_file_dir) and ("test" in output_file_dir):
        model_type = "test"
      elif ("validation" in input_file_dir) and ("validation" in output_file_dir):
        model_type = "validation"

      with open(output_file_dir + '/' + name + '_spect_' + model_type + '.npy', 'wb') as f:
          np.save(f, np.array(bird_spect_lst))
            
      temp = [name] * len(bird_spect_lst) 
      count_data += temp

    spect_data["spect"] = np.array(spect_lst) 
    spect_data["shape"] = spect_data["spect"].shape
    return spect_data, count_data
```

```python
test_arrays = '../bird-sounds-data/validation/arrays'
test_spectrograms = '../bird-sounds-data/validation/spectrograms'

test_dict, test_data = get_melspectrogram(test_arrays, test_spectrograms)
x_test = test_dict["spect"]
print(test_dict["shape"])
print(len(test_data))
print('Test data gathered')
```

```python
def get_y(clusters, bird_lst,predict_data):
    """This function y_true values"""
    y = [] 
    
    for idx,ele in enumerate(clusters):    
        item_name = predict_data[idx]
        lbl = bird_lst.index(item_name)
        y.append(lbl)
    return y 
```

```python
y_pred = clusters

bird_lst = ['asian_koel', 'blue_jay', 'cockatoo', 'crane', 'crow', 'cuckoo', 'dove', 'duck', 'eagle', 'falcon', 
            'goose', 'grackles', 'hawk', 'hen', 'humming_bird', 'kingfisher', 'loon', 'ostrich', 'parrot', 'penguin',
            'pigeon', 'quail', 'red_cardinal', 'seagull', 'sparrow', 'starling', 'swan', 'tufted_titmouse', 'turkey', 'wood_thrush']

y = get_y(clusters,bird_lst,test_data)
```

```python
from sklearn.utils.linear_assignment_ import linear_assignment

def class_assigner(y_true, y_pred):
    """This function uses the Hungarian Algorithm to assign classes to the clustering labels"""
    new_bird_lst = []
    y_true = y_true.astype(np.int64)
    assert y_pred.size == y_true.size
    D = max(y_pred.max(), y_true.max()) + 1
    w = np.zeros((D, D), dtype=np.int64)
    for i in range(y_pred.size):
        w[y_pred[i], y_true[i]] += 1
    ind = linear_assignment(w.max() - w)
    for m in ind:
        new_bird_lst.append(bird_lst[m[1]])
    return new_bird_lst 
```

```python
y = np.array(y)
y_pred = np.array(y_pred)
new_bird_lst = class_assigner(y,y_pred)
```
